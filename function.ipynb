```python
def cal(a,b,o):
    if o == '+':
        print(a+b)
    elif o == '-':
        print(a-b)
    elif o == '*':
        print(a*b)
    elif o =='/':
        print(a/b)

    else:
        print("Invalid operator")
     
    
a = int(input("Enter a = "))
b = int(input("Enter  b = "))
o = input("Enter + - * / ")
cal(a,b,o)
```

```python
def cal(l,b):
    a = l*b           #local variable
    return a
    
    
l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))

x = cal(l,b)              #argument
print("The area is = ",x)
v = x*h
print("The volume = ",v)
```

```python
def cal(l,b,h):
    a = l*b
    v = a*h
    
    return a,v

area,volume= cal(10,5,2)
print("The area is = ",area)
print("The volume is = ",volume)
```

```python
net_total = 0
def cal():
    global net_total
    name = input("Enter name = ")
    price = int(input("Enter phone = "))
    quantity = int(input("Enter quantity = "))
    total = price*quantity
    net_total = net_total+total
#     s = s+name+" "+str(price) + " "+str(quantity) + " "+ str(total)+"\n"
    print(name,price,quantity,total)
    x = input("Enter space for more entry= ")
    if x == ' ':
        cal()

cal()
print(net_total)
```
