```python
s = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    phone = int(input("Enter phone = "))
    
    s = s+f"{name},{phone} \n"
print(s)

b = open('phone.csv','a')
b.write(s)
b.close()
```

```python
s = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter address = ")
    phone = int(input("Enter phone = "))
    
    s = s+f"{name},{age},{add},{phone} \n"
print(s)

b = open('data.csv','w')
b.write(s)
b.close()
```
